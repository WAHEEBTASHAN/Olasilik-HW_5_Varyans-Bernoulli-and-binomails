**Variance**

The Variance is defined as: The average of the squared differences from the Mean.

To calculate the variance  we need to follow these steps:

1- We need to find out the mean ( the simple average of the numbers)

2- Then for each number: subtract the Mean and square the result (the squared difference)

3- Then work out the average of those squared differences.
![alt text](https://github.com/WAHEEBTASHAN/Olasilik-HW_5_Varyans-Bernoulli-and-binomails/blob/master/variance%20picture.JPG?raw=true)

**Bernoulli**

Bernoulli trials are such trials which has only two possible outcomes. For example, tossing

a coin will give only two outputs, head or tail. Such a trial is known as Bernoulli trial. If

probability of one of the event will be p then the probability of the other event will be (1 - p).



**Example**

A basketball player takes four independent free throws with a probability of 0.7 of getting a

basket on each shot. The number of baskets made is recorded. find the Bernoulli 

experiment with 4 trials shown in this example 

**Solution** 

The probability of success is p = 0.7 and the probability of failure is q = 1 − p = 0.3.

How ever We are 

interested in the variable X which counts the number of successes in 4 trials. This is

an example of a Bernoulli experiment with 4 trials.

E[X]= ∑ xP(X = x) = 0(1-p) + 1p = p

E[X^2]= ∑ x^2*P(X = x^2) = 0^2(1-p) + 1^2*p = p

Var(X) = E[X^2]-E[X]^2 = p - p^2 = p(1-p) = p(1-p)
