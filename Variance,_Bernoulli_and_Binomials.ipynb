**1- Variance**

The Variance is defined as: The average of the squared differences from the Mean.

To calculate the variance  we need to follow these steps:

1- We need to find out the mean ( the simple average of the numbers)

2- Then for each number: subtract the Mean and square the result (the squared difference)

3- Then work out the average of those squared differences.
![alt text](https://github.com/WAHEEBTASHAN/Olasilik-HW_5_Varyans-Bernoulli-and-binomails/blob/master/variance%20picture.JPG?raw=true)

**2- Bernoulli**

Bernoulli trials are such trials which has only two possible outcomes. For example, tossing

a coin will give only two outputs, head or tail. Such a trial is known as Bernoulli trial. If

probability of one of the event will be p then the probability of the other event will be (1 - p).

The probability function of binomial distribution is:

P(X=x)=(p^x)*(1-p)^1-x & x(0,1)

```
# Solving Example 1 using python 

hights = [600, 470, 170, 430, 300]
#First  we need to print the hights
def print_hights(hights):
    for hight in hights:
        print(hights)

#calculate the sum
def hights_sum(hights):
    total = 0
    for hight in hights:
        total += hight
    return total

#Take the average by adding up each grade
def hights_average(hights):
    sum_of_hights = hights_sum(hights)
    average = sum_of_hights / float(len(hights))
    return average

#Here we calculate the variance given the formula above
#by subtracting each score from the average and then taking square and rolling the sum.
def hights_variance(scores):
    average = hights_average(scores)
    variance = 0
    for score in scores:
        variance = variance + (average - score) ** 2

    return variance/len(scores)

print('the variance is: ',hights_variance(hights))

#And here the standard deviation comes by taking square root of the variance
def hights_std_deviation(variance):
    return variance ** 0.5
variance = hights_variance(hights)

print('the standard deviation is: ',hights_std_deviation(variance))
```
